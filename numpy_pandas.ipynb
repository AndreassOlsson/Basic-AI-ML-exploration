```
# Get populartySong, artistName, populartyArtist as np arrays
trackRating = np.array(dT["populartySong"])
artist = np.array(dT["artistName"])
artistRating = np.array(dT["populartyArtist"])
```

```
# Get mean of np arrays
tR_mean = np.mean(trackRating)
aR_mean = np.mean(artistRating)
```

```
# Create a numpy array with artists whose pop rating >= 70
tA_df = dT.loc[dT['populartyArtist'] >= 70]
topArtists = sorted(set(np.array(tA_df['artistName'])))
topArtists
```

```
# Good tracks by good artists
tT_df = dT.loc[dT['populartySong'] >= 70]
topTracks = sorted(set(np.array(tT_df['trackName'])))

topArtists_tracks = sorted(set(np.array(tA_df['trackName'])))

topTracks_by_topArtists = np.intersect1d(topTracks, topArtists_tracks)
topTracks_by_topArtists
```

```
# Trackratings for every top artist
for artist in topArtists:
  allTracks = dT.loc[dT['artistName'] == artist]

  # Above 70 rating
  artist_topTracks = allTracks.loc[allTracks['populartySong'] >= 70]

  # Mean of tracks
  artist_tracks_mean = np.mean(allTracks['populartySong'])

  print(f'{artist}, No. top tracks: {len(artist_topTracks)}, Mean track rating: {artist_tracks_mean}\n')

```

```
tD = pd.read_csv('/content/drive/MyDrive/Assignments/1/TracksDetail.csv')
tD
```

```
# Remove first column
tD.drop('Unnamed: 0', axis=1, inplace=True)
```

```
# Remove artistName and trackName columns
tD.drop(['artistName', 'trackName'], axis=1, inplace=True)
```

```
# Rename albumName to track
tD.rename(columns={'albumName':'track'}, inplace=True)
```

```
# Merge Tracks and TrackDetails on track column
merged_df = pd.merge(t, tD, on='track', how='outer')
merged_df
```

```
# List the 10 most popular genres and how many songs they include
s = merged_df['genres'].value_counts()
mpg_count = s[:10]
mpg_count
```

```
# List all the tracks and their artists in the “electropop“ genre
ePop = merged_df.loc[merged_df['genres'].str.contains('electropop')]
ePop
```
